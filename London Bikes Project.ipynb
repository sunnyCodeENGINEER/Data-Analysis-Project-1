```python
my_col_names_dict = {
    "timestamp": "time",
    "cnt": "count",
    "t1": "temp_real_C",
    "t2": "temp_feels_like_C",
    "hum": "humidity_percentage",
    "windspeed": "windspeed_kph",
    "weather_code": "weather",
    "is_holiday": "is_holiday",
    "is_weekend": "is_weekend",
    "season": "season"
}

# Renaming columns
data.rename(my_col_names_dict, axis=1, inplace=True)
```

```python
data.head()
```

```python
data.season.value_counts()
```

```python
data.humidity_percentage = data.humidity_percentage / 100

season_dict = {
    "0.0": "spring",
    "1.0": "summer",
    "2.0": "autumn",
    "3.0": "winter"
}

weather_dict = {
    "1.0": "Clear",
    "2.0": "Scattered clouds",
    "3.0": "Broken clouds",
    "4.0": "Cloudy",
    "7.0": "Rain",
    "10.0": "Rain with thunderclouds",
    "26.0": "Snowfall"
}

data.season = data.season.astype("str")
data.weather = data.weather.astype("str")

data.season = data.season.map(season_dict)
data.weather = data.weather.map(weather_dict)
```

```python
data.head()
```

```python
boolean_dict = {
    "1.0": "True",
    "0.0": "False"
}

data.is_holiday = data.is_holiday.astype("str")
data.is_weekend = data.is_weekend. astype("str")

data.is_holiday = data.is_holiday.map(boolean_dict)
data.is_weekend = data.is_weekend.map(boolean_dict)
```

```python
data.head()
```

```python
# Export cleaned up data to an excel file
data.to_excel("london_bikes_final.xlsx", sheet_name="Data")
```
